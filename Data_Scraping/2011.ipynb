```python
import pandas as pd
import numpy as np
```

```python
avg_drivingdistance = pd.read_html('https://www.pgatour.com/content/pgatour/stats/stat.101.y2011.eon.t014.html')
drivingaccuracy = pd.read_html('https://www.pgatour.com/content/pgatour/stats/stat.102.y2011.eon.t014.html')
gir = pd.read_html('https://www.pgatour.com/content/pgatour/stats/stat.103.y2011.eon.t014.html')
scrambling = pd.read_html('https://www.pgatour.com/content/pgatour/stats/stat.130.y2011.eon.t014.html')
sandsaves = pd.read_html('https://www.pgatour.com/content/pgatour/stats/stat.111.y2011.eon.t014.html')
putting_avg = pd.read_html('https://www.pgatour.com/content/pgatour/stats/stat.104.y2011.eon.t014.html')
```

```python
df1 = avg_drivingdistance[1][['PLAYER NAME','AVG.']].rename(columns={'AVG.': 'avg_driv_dist'}).set_index('PLAYER NAME')
```

```python
df2 = drivingaccuracy[1][['PLAYER NAME','%']].rename(columns={'%': 'driv_acc%'}).set_index('PLAYER NAME')
```

```python
df3 = gir[1][['PLAYER NAME','%']].rename(columns={'%': 'gir%'}).set_index('PLAYER NAME')
```

```python
df4 = scrambling[1][['PLAYER NAME','%']].rename(columns={'%': 'scrambling%'}).set_index('PLAYER NAME')
```

```python
df5 = sandsaves[1][['PLAYER NAME','%']].rename(columns={'%': 'sandsaves'}).set_index('PLAYER NAME')
```

```python
df6 = putting_avg[1][['PLAYER NAME','AVG']].rename(columns={'AVG': 'putting_avg'}).set_index('PLAYER NAME')
```

```python
leaderboard = pd.read_html('https://www.espn.com/golf/leaderboard/_/tournamentId/980')[0]
leaderboard['PLAYER'] = leaderboard['PLAYER'].str.replace(r"\(.*\)","").str.replace(r'[^a-zA-Z .]+', '').str.strip()
leaderboard = leaderboard[['PLAYER','TO PAR']].rename(columns={'TO PAR': 'score','PLAYER': 'PLAYER NAME'}).set_index('PLAYER NAME')
leaderboard
```

```python
merged_df = pd.concat([df1, df2, df3, df4, df5, df6, leaderboard], axis=1, sort=False)
```

```python
merged_df.isnull().sum()
```

```python
merged_df[pd.isnull(merged_df).any(axis=1)]
```

```python
merged_df = merged_df[~pd.isnull(merged_df).any(axis=1)]
merged_df['year'] = '2011'
merged_df
```

```python
merged_df.to_csv('2011.csv')
```
